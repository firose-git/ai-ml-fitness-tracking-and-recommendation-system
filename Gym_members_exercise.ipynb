```python
#Processing Data
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
```

```python
obj = gym_data.select_dtypes(include='object')
non_obj = gym_data.select_dtypes(exclude='object')
```

```python
for i in range(0 , obj.shape[1]):
    obj.iloc[:,i] = lab.fit_transform(obj.iloc[:,i])
```

```python
gym_data =pd.concat([obj , non_obj] , axis=1)
```

```python
gym_data
```

```python
x = gym_data.drop(['Experience_Level'],axis=1)
y = gym_data.Experience_Level
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test , y_pred)
print('Accuracy:' , accuracy)
```

```python
model = SVC()
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
print('Accuracy:' , accuracy)
```

```python
model = KNeighborsClassifier(n_neighbors=7)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
print('Accuracy:' , accuracy)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
print('GausianNB Accuracy:' , accuracy)
```

```python
model = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
print('Accuracy:' , accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred_dt = model.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
models = {
    "Decision Tree": accuracy_score(y_test, y_pred_dt),
    "Random Forest": accuracy_score(y_test, y_pred_rf),
}

for name, acc in models.items():
    print(f"{name} Accuracy: {acc:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
import matplotlib.pyplot as plt

model_names = ['Decision Tree', 'Random Forest', 'Logistic Regression', 'KNN']
accuracies = [
    accuracy_score(y_test, y_pred_dt),
    accuracy_score(y_test, y_pred_rf),
    accuracy_score(y_test, y_pred_lr),
    accuracy_score(y_test, y_pred_knn)
]

plt.figure(figsize=(8,5))
plt.bar(model_names, accuracies, color='skyblue')
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.ylim(0, 1)
plt.show()
```
